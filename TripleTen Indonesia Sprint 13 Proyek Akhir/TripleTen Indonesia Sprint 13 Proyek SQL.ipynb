## Tasks:

### Task 1: Hitung jumlah buku yang dirilis setelah tanggal 1 Januari 2000.

```python
books_2000 = pd.io.sql.read_sql(
    '''
    SELECT COUNT(DISTINCT book_id) 
    FROM books 
    WHERE publication_date > '2000-01-01' 
    ''', con = engine)

display(books_2000)
```

**Kesimpulan:**

* Jumlah buku yang dirilis setelah tanggal 1 Januari 2000 adalah sebanyak **819** buku.

### Task 2: Hitung jumlah ulasan pengguna dan rating rata-rata untuk setiap buku.

```python
query = ''' 
        SELECT 
            books.book_id, 
            books.title, 
            sub.avg_rating,
            sub.user_reviews
        FROM 
            (SELECT
                reviews.book_id AS book_id,
                COUNT(DISTINCT reviews.review_id) AS user_reviews,
                AVG(ratings.rating) AS avg_rating
             FROM
                reviews
                INNER JOIN ratings ON ratings.book_Id = reviews.book_id
             GROUP BY
                 reviews.book_id
             ORDER BY
                 user_reviews DESC,
                 avg_rating DESC) AS sub
            INNER JOIN books ON sub.book_id = books.book_id
        ORDER BY 
            user_reviews DESC,
            avg_rating DESC
        '''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

**Kesimpulan:**

* Dari data yang kami analisis, terdapat 994 buku yang memiliki informasi tentang jumlah ulasan pengguna dan rating rata-rata
* Jumlah ulasan tertinggi ditemukan pada buku dengan judul Twilight.

### Task 3: Identifikasi penerbit yang telah menerbitkan jumlah buku terbanyak, dengan jumlah halaman lebih dari 50.

```python
query = ''' 
        SELECT 
            publishers.publisher, 
            COUNT(DISTINCT books.book_id) AS no_of_books
        FROM 
            publishers
        INNER JOIN books ON books.publisher_id = publishers.publisher_id
        WHERE 
            books.num_pages > 50 
        GROUP BY 
            publishers.publisher
        ORDER BY 
            no_of_books DESC
        LIMIT 10
        '''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

**Kesimpulan:**

* Dalam tabel tersebut, kami dapat melihat penerbit yang mendominasi dalam menerbitkan buku dengan lebih dari 50 halaman.
* Informasi yang kami miliki menunjukkan bahwa penerbit Penguin Books telah menerbitkan jumlah buku terbanyak, mencapai angka 42 buku

### Task 4: Identifikasi penulis dengan rating rata-rata buku tertinggi: temukan buku dengan rating minimal 50.

```python
query = ''' 
        SELECT 
            authors.author AS author_name,
            AVG(ratings.rating) AS avg_rating
        FROM 
            authors
        INNER JOIN books ON books.author_id = authors.author_id
        INNER JOIN ratings ON ratings.book_id = books.book_id
        GROUP BY 
            author_name
        HAVING COUNT(ratings.rating) > 50
        ORDER BY 
            avg_rating DESC
        '''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

**Kesimpulan:**

* Dalam tabel yang disajikan, terdapat daftar penulis yang memiliki peringkat buku rata-rata tertinggi. Namun, hanya buku-buku dengan setidaknya 50 peringkat yang masuk dalam perhitungan ini.
* Dalam data yang kami tinjau, ada sebanyak 23 penulis yang memiliki buku dengan rating minimal sebesar 50.
* Menurut data yang kami miliki, penulis J.K. Rowling/Mary GrandPré memiliki peringkat buku tertinggi.

### Task 5: Hitung jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku.

```python
query = '''
SELECT
     SUM(Sub.count_reviews)/COUNT(Sub.count_reviews) AS count

FROM(    
    select reviews.username, 
    count(reviews.username) as count_reviews
    from reviews
    where reviews.username in (
                                select ratings.username
                                from ratings
                                group by ratings.username
                                having count(ratings.rating_id) > 50
    
                                )
    group by username) as Sub

'''

result = pd.io.sql.read_sql(query, con = engine)
display(result)
```

**Kesimpulan:**

* Pengguna yang memberi rating terhadap lebih dari 50 buku, rata-rata menulis 24.333333 ulasan per pengguna.
